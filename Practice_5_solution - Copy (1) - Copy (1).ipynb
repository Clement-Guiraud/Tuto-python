First let's create a function that will open and read files and return a dictionary with all the content of the data, and as keys the ids of the tweets listed. 


```python
import csv
def readfiles(file):
    '''This function reads structured tweet files and returns a dictionary 
    with tweet id considered as keys of the output '''
    
    # Creating an empty dictionary 
    dict_tweets = {}
    err_msg = 'Error: Cannot find the file'

    try: 
        with open(file,'r') as handle:
            content = csv.reader(handle)
            # This to squeeze the first line 
            next(content)
            for line in content: 
                dict_tweets[line[0]] = line[1:]
    except: 
        print(err_msg)
        return
    
    # As you can see here, we do not need a print output
    # as we are going to use this function inside a global program.
    # So we do not want to pollute the output displayed informations
    # given to the users 
    return dict_tweets

help(readfiles)
```

Now let's create a function which focuses on the content of the tweets.

Create a function that takes a dictionary as argument and returns a dictionary in which you will find globally extract the different words used by Trump & Biden and count the number of times they are used.


```python
def count_words(dictionary,index): 
    '''This function takes a dictionary and an index element as argument.
    The dictionary is composed of id, timestamp, tweet content, retweets, likes, ...
    The index indicates where the content of the tweets are located in your csv files.
    The function returns a new dictionary in which 
    keys are each word used in the tweets read and counts how many times they appear.'''
    
    dict_words = dict()
    err_msg = 'Error: Your input is not a dictionary'
    
    if isinstance(dictionary,dict)==False :    
        print(err_msg)
        return None
    
    for keys,values in dictionary.items():
        tweets = values[index].split(" ")
        for word in tweets:
            word = word.lower()
            if word in dict_words: 
                dict_words[word] += 1
            else :
                dict_words[word] = 1

    return dict_words
    
```

```python
# See here some details about stopwords and nltk 
import string 
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')

def count_words(dictionary,index): 
    '''This function takes a dictionary and an index element as argument.
    The dictionary is composed of id, timestamp, tweet content, retweets, likes, ...
    The index indicates where the content of the tweets are located in your csv files.
    The function returns a new dictionary in which 
    keys are each word used in the tweets read and counts how many times they appear.'''
    
    dict_words = dict()
    err_msg = 'Error: Your input is not a dictionary'
    
    if isinstance(dictionary,dict)==False :    
        print(err_msg)
        return None
    
    for keys in dictionary:         
        # Using index here to extract the content of tweets only       
            
        tweets = dictionary[keys][index].split(" ")
        punct = [s for s in string.punctuation]
        for word in tweets:
            # Excluding non useful words (what we call stop words)
            # But of course we have to exclude when they are linked with 
            # some punctuations also ...
            word = word.lower().translate(str.maketrans('', '', string.punctuation))
            if word in punct: 
                continue
            if word in stopwords.words('english'):
                continue
            if word in dict_words: 
                dict_words[word] += 1
            else :
                dict_words[word] = 1
          
    return dict_words
```
