```python
def extract_fewdata(file,index):

    error_msg = 'The index argument is not a list'
    
    # BEGIN SOLUTION
    if type(index)!= list:
        print(error_msg)
        return None
    
    data = []
    for i in index: 
        data.append(extract_data(file,i))
        
    return data
    # END SOLUTION
```

If the file is empty, return None and print out an "warning" message.


```python
def extract_fulldata(file):
    '''This function takes 1 argument which corresponds 
    to the csv file path. The function creates one main list 
    containing all extracted data, separated in multiple lists. 
    '''
    
    warning_msg = "Warning: The given csv file is empty!"
    
    #BEGIN SOLUTION
    ## Initialisation:
    with open(file,'r', encoding='utf-8') as handle:
        for i in handle:
            length = len(i.split(','))
            break
        index = list(range(length-1))
    
    return extract_fewdata(file,index)
    #END SOLUTION
```
