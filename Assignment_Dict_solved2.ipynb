### Working with a NBA dataset 

Back to our NBA dataset.

You will read et extract data from the file and manipulate them but this time using dictionaries.

The exercise here consists in creating a function that :
- take a file as argument
- build up a dictionary containing all data from the file

    The keys of the dictionary should be the title of columns. 

    And value for each key should be a list of data from the same column.


**Note: 
The main function should be call dict_data(file) !**

**Tip: 
I advise you to consider multiple functions (as we did in the last assignment). But it can also be done with one (with more steps and less "pythonish")**

```python
import csv 

err_msg = "Error: the file has not be found"
def extract_data(file,index):
    datalist = []
    try:
        with open (file, 'r') as handle:
            csv_reader = csv.reader(handle)
            for row in csv_reader:
                datalist.append(row[index])
    except:
        print(err_msg)
        return 
    numbers_list = []
    for i in datalist[1:]:
        try:
            numbers_list.append(float(i))
        except:
            numbers_list.append(i)
    return numbers_list


def dict_data(file):
    '''This function takes a csv file as argument.
    It extracts data from the NBA dataset file,
    And compiles all of them into a dictionary.
    The keys of this one corresponds to column titles.
    '''
        
    ## Initialisation:
    data = {}
    err_msg = "Error: Your file doesn't exist nor a csv format"
    
    try:
        with open(file,'r', encoding='utf-8') as handle:
            content = csv.reader(handle)
            # Reading and saving only the first row 
            # (=column titles)
            titles = next(content)
#             for row in content:
#                 titles = row
#                 break
    except: 
        print(err_msg)
        return None
        
    # Exploring each column and extracting data from those
    # to put them into a structure dictionary
    index = list(range(len(titles)-1))    
    for i in index: 
        data[titles[i]] = extract_data(file,i)
                   
    return data
```

### Working with a space mission dataset 

This is a very specific function to be programmed, but it will allow you to train on using loops, lists and dictionary.

We want to create a function that takes one argument which is a CSV file (here it only works with SpaceMission_data.csv file). 

The function explores the file and returns a dictionary in which it counts :
- The number of mission for each company (company will be keys of the dictionary)
- The number of successful missions 
- The number of failed missions

Each value element of the dictionary will be a list of the 3 counts described above. 
It will look like this: 
{company : [total number of missions, number of successful missions, number of failed missions]}

Explore the file first, and find a solution to identify the successful and failed missions identification. 

You will see that it is not only a question about a "," delimiter. 
So in order to handle easily the reading of the format, i strongly suggest you to use this method: 
Just after opening the file, 
use "content = csv.reader(handle)"

And you will be able to loop into content.

And of course return a dictionary! 

```python
import csv
def space_statistics(filename):
    '''This function takes one argument which should be a CSV file. 
    It specifically explores the space_corrected.csv file. 
    The function returns a dictionary in which it counts the number of mission for each company in charge
    and also counts the number of successful and failed missions'''

    space = dict()
    err_Msg = 'Error: Check the file path or content' 
 
    #BEGIN SOLUTION 
    try: 
        with open(filename,'r') as handle:
            content = csv.reader(handle)
            next(content)
                        
            for line in content: 
                                
                # Extracting the company name
                company = line[2]
                # Checking if the company name has already an entry in the dictionary
                if company not in space: 
                    # Counting if it is a success or kind of a failure (partial, or other)
                    if 'failure' in line[8].lower():
                        space[company]=[1,0,1]
                    elif 'success' in line[8].lower():
                        space[company]=[1,1,0]
                        
                else: 
                    if 'failure' in line[8].lower():
                        success_list = space[company]
                        success_list[0] += 1
                        success_list[2] += 1
                        space[company]= success_list
                    elif 'success' in line[8].lower():
                        success_list = space[company]
                        success_list[0] += 1
                        success_list[1] += 1
                        space[company]= success_list
    except: 
        print(err_msg)
        return None

    return space

    #END SOLUTION 

help(space_statistics)
```
